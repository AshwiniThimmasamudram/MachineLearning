# encoding binary variables



```python
df['fueltype'].unique()
```

```python
# encoding FUEL TYPE 
# 1 corresponds to gas
# 0 corresponds to diesel

def binary_map(x):
    return x.map({'gas':1, 'diesel':0})
df[['fueltype']]=df[['fueltype']].apply(binary_map)
df.head()
```

```python
df['aspiration'].unique()
```

```python
# encoding ASPIRATION 
# 1 corresponds to std
# 0 corresponds to turbo

def binary_map(y):
    return y.map({'std':1, 'turbo':0})
df[['aspiration']]=df[['aspiration']].apply(binary_map)
df.head()
```

```python
df['doornumber'].unique()
```

```python
# encoding DOOR NUMBER 
# 1 corresponds to two
# 0 corresponds to four

def binary_map(x):
    return x.map({'two':1, 'four':0})
df[['doornumber']]=df[['doornumber']].apply(binary_map)
df.head()
```

```python
df['enginelocation'].unique()
```

```python
# encoding ENGINE LOCATION
# 1 corresponds to FRONT
# 0 corresponds to REAR


def binary_map(x):
    return x.map({'front':1, 'rear':0})
df[['enginelocation']]=df[['enginelocation']].apply(binary_map)
df.head()
```

# creating dummy variables to categorical variabels

```python
df['carbody'].unique()
```

```python
# encoding CAR BODY
# 0000 corresponds to convertible
# 1000 corresponds to hardtop
# 0100 corresponds to hatchback
# 0010 corresponds to sedan
# 0001 corresponds to wagon


car_body = pd.get_dummies(df['carbody'])
car_body = pd.get_dummies(df['carbody'], drop_first =  True)
df=pd.concat([df,car_body],axis=1)
df.head()
```

```python
# dropping old column name 'carbody' after encoding
df.drop(columns =["carbody"], inplace = True)
df.head()
```

```python
df['enginetype'].unique()
```

```python
# encoding Enginetype column
# 000000 corresponds to 
# 100000 corresponds to dohcv
# 010000 corresponds to I
# 001000 corresponds to ohc
# 000100 corresponds to ohcf
# 000010 corresponds to ohcv
# 000001 corresponds to rotor

e_type = pd.get_dummies(df['enginetype'])
e_type = pd.get_dummies(df['enginetype'], drop_first =  True)
df=pd.concat([df,e_type],axis=1)
df.head()
```

```python
# droping engine type column
df.drop(columns =["enginetype"], inplace = True)
df.head()
#df.enginetype
```

```python
# droping DRIVEWHEEL type column
df.drop(columns =["drivewheel"], inplace = True)
df.head()
#df.drivewheel
```

# Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (24,15))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
# importing minmaxscalar package from sklearn library
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df.info()
```

# applying minmax scaling on numerical variables

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
number_variables = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'boreratio','stroke','compressionratio','price']

df_train[number_variables] = scaler.fit_transform(df_train[number_variables])

df_train.head()
```

```python
df_train.describe()
```

```python
# buliding model and removing target variable from y-train data set
y_train = df_train.pop('price')
X_train = df_train
```

# Buliding Linear Model

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```
